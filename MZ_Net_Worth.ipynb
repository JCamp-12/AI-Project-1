```
import pandas as pd
import matplotlib.pyplot as plt

# Sample data for Mark Zuckerberg's net worth over years
data = {
    'Year': [2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'Net Worth ($B)': [44.6, 56.0, 71.0, 62.3, 54.7, 97.0, 67.3, 115.0, 165]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Net Worth ($B)'], marker='o', linestyle='-', color='b')
plt.title("Mark Zuckerberg's Net Worth Over Time")
plt.xlabel('Year')
plt.ylabel('Net Worth ($B)')
plt.grid(True)
plt.show()
```
