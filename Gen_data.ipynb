```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Using MatplotLib to Chart a DataFrame

```python
# Load in csv
gen_df = pd.read_csv("./resources-AF/dfa_generation_shares.csv")
gen_df.head()
```

```python
#Trying to make 'Date' column into a datetime data type
gen_df['Date'] = pd.to_datetime(gen_df['Date'], format='%Y:Q%q')
```

```python
gen_df[["Date", "Category", "Net worth", "Assets", "Real estate"]]
gen_df.head()
```

```python
#Removing the 'silent' in category.
filtered_gen_df = gen_df[gen_df["Category"] != "Silent"]
filtered_gen_df.head()
```

```python
multi_gen_df = filtered_gen_df.set_index(["Date"])
multi_gen_df.head()
```

```python
# Making 'Date' the index.
final_gen_df = multi_gen_df[["Category", "Net worth", "Assets", "Real estate"]]
final_gen_df.head() 
```

```python
#Finding the average for each column grouped by generation. Resetting the index inpreperation for bar graph. 
grouped_final_df = final_gen_df.groupby('Category')[['Net worth', 'Assets', 'Real estate']].mean().reset_index()
grouped_final_df
```

```python
#Bar graph showing the mean of each generation in Networth, Assets, and Real estate. This is not by grouped by age 35. 
plt.figure(figsize=(10,6))

bars = plt.bar(grouped_final_df['Category'], grouped_final_df['Net worth'], width=0.5)
bars = plt.bar(grouped_final_df['Category'], grouped_final_df['Assets'], width=0.5, bottom=grouped_final_df['Net worth'])
bars = plt.bar(grouped_final_df['Category'], grouped_final_df['Real estate'], width=0.5, bottom=grouped_final_df['Assets']+grouped_final_df['Net worth'])

plt.xlabel('Category', fontsize=14)
plt.ylabel('Mean Value', fontsize=14)
plt.xticks(grouped_final_df['Category'], rotation=45)
plt.legend(['Net worth', 'Assets', 'Real Estate'])
plt.title('Mean Net worth, Assets, and Real Estate by Generation', fontsize=16)

plt.show()

```

```python
grouped_final_df
```

```python
#Trying to see how many of the assets are Real estate; calculating the ratio/percentage of the 'real estate' column value divided by the 'Assests' column value
ratio_by_gen = grouped_final_df.groupby('Category')[['Real estate', 'Assets']].apply(lambda x: x['Real estate'].mean() / x['Assets'].mean())
ratio_by_gen = ratio_by_gen.reset_index()
ratio_by_gen
```

```python
#First test line graph 
fig, ax = plt.subplots(figsize=(30, 6))

for category in final_gen_df['Category'].unique():
    category_data = final_gen_df[final_gen_df['Category'] == category]
    ax.plot(category_data.index, category_data['Net worth'], label=category + ' (Net worth)')
    ax.plot(category_data.index, category_data['Assets'], linestyle='--', label=category + ' (Assets)')
    ax.plot(category_data.index, category_data['Real estate'], linestyle=':', label=category + ' (Real estate)')

ax.set_title('Net Worth, Assets, and Real Estate by Generation')
ax.set_xlabel('Date')
ax.legend()

plt.show()
```
