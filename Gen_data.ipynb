```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Using MatplotLib to Chart a DataFrame

```python
# Load in csv
gen_df = pd.read_csv("./resources-AF/dfa_generation_shares.csv")
gen_df.head()
```

```python
#Trying to make 'Date' column into a datetime data type
gen_df['Date'] = pd.to_datetime(gen_df['Date'], format='%Y:Q%m')

gen_df['Year'] = gen_df['Date'].dt.year
gen_df
```

```python
gen_df[["Category", "Net worth", "Assets", "Real estate"]]
gen_df.head()
```

```python
#Removing the 'silent' in category.
filtered_gen_df = gen_df[gen_df["Category"] != "Silent"]
filtered_gen_df.head()
```

```python
multi_gen_df = filtered_gen_df.set_index(["Year"])
multi_gen_df.head()
```

```python
# Making 'Year' the index.
final_gen_df = multi_gen_df[["Category", "Net worth", "Assets", "Real estate"]].reset_index('Year')
final_gen_df.head() 
```

```python
grouped_final_df = final_gen_df.groupby(['Year', 'Category'])[['Net worth', 'Assets', 'Real estate']].mean()
grouped_final_df.head(50)
```

```python
#grouped_final_df = grouped_final_df.reset_index()
#grouped_final_df
```

```python
# Define the generations to include
generations = ['BabyBoom', 'GenX', 'Millennial']

# Filter the data to include only the years from 1989 to 2024 and the specified generations
filtered_data = grouped_final_df.loc[(grouped_final_df.index.get_level_values('Year') >= 1989) & 
                                              (grouped_final_df.index.get_level_values('Year') <= 2024) &
                                              (grouped_final_df.index.get_level_values('Category').isin(generations))]

# Reset index to use 'Year' as x-axis
filtered_data_reset = filtered_data.reset_index()

# Create a line chart for each generation and each category
for generation in generations:
    generation_data = filtered_data_reset[filtered_data_reset['Category'] == generation]
    plt.plot(generation_data['Year'], generation_data['Assets'], label=f'{generation} Assets')
    plt.plot(generation_data['Year'], generation_data['Net worth'], label=f'{generation} Net Worth')
    plt.plot(generation_data['Year'], generation_data['Real estate'], label=f'{generation} Real Estate')

# Add labels, title, and legend
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Assets, Net Worth, and Real Estate by Generation from 1989 to 2024')
plt.legend()

# Display the plot
plt.show()
```

```python
# Define the generations to include
generations = ['BabyBoom', 'GenX', 'Millennial']

# Filter the data to include only the years from 1989 to 2024 and the specified generations
filtered_data = grouped_final_df.loc[(grouped_final_df.index.get_level_values('Year') >= 1989) &
                                     (grouped_final_df.index.get_level_values('Year') <= 2024) &
                                     (grouped_final_df.index.get_level_values('Category').isin(generations))]

# Reset index to use 'Year' as x-axis
filtered_data_reset = filtered_data.reset_index()

# Create a figure with subplots
fig, axes = plt.subplots(nrows=3, figsize=(12, 10), sharex=True)

# Create a line chart for each generation and each category
for i, metric in enumerate(['Assets', 'Net worth', 'Real estate']):
    ax = axes[i]
    for generation in generations:
        generation_data = filtered_data_reset[filtered_data_reset['Category'] == generation]
        ax.plot(generation_data['Year'], generation_data[metric], label=f'{generation} {metric}')
    ax.set_ylabel(metric)
    ax.legend()

# Set the x-axis limits and label
axes[2].set_xlim(1989, 2024)
axes[2].set_xlabel('Year')

# Adjust spacing between subplots
fig.subplots_adjust(hspace=0.5)

# Display the plot
plt.show()
```
