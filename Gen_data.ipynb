```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Using MatplotLib to Chart a DataFrame

```python
# Load in csv
gen_df = pd.read_csv("./resources-AF/dfa_generation_shares.csv")
gen_df.head()
```

```python
gen_df[["Date", "Category", "Net worth", "Assets", "Real estate"]]
gen_df.head()
```

```python
multi_gen_df = gen_df.set_index(["Date"])
multi_gen_df.head()
```

```python
final_gen_df = multi_gen_df[["Category", "Net worth", "Assets", "Real estate"]]
final_gen_df.head() 
```

```python
final_gen_df.plot()
```

```python
fig, ax = plt.subplots(figsize=(20, 6))

for category in final_gen_df['Category'].unique():
    category_data = final_gen_df[final_gen_df['Category'] == category]
    ax.plot(category_data.index, category_data['Net worth'], label=category + ' (Net worth)')
    #ax.plot(category_data.index, category_data['Assets'], linestyle='--', label=category + ' (Assets)')
    #ax.plot(category_data.index, category_data['Real estate'], linestyle=':', label=category + ' (Real estate)')

ax.set_title('Net Worth, Assets, and Real Estate by Generation')
ax.set_xlabel('Date')
ax.legend()

plt.show()
```
